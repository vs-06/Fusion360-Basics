```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('world_population.csv')
df.head(10)
```

```python
df[df['Rank'] <= 10]
```

```python
asian_countries = df[df['Country'].isin(['India', 'Pakistan', 'China'])]
display(asian_countries)

united_countries = df[df['Country'].str.contains('United')]
united_countries
```

# FILTERING DATA USING {.filter() and iems = [] and like = } FOR COLUMN AND ROW WISE FILTRATION

```python
df2 = df.set_index('Country')
df2.head()

display(df2.filter(items = ['Continent', 'CCA3', 'Capital']))

df2[['Continent', 'CCA3', 'Capital']]

```

```python
# FILTERING ROWS USING LIKE AND ITEMS AND AXIS = 0

display(df2.filter(items = ['India'], axis = 0))

df2.filter(like ='United', axis = 0)
```

 # LOC and ILOC - For filtering data "CELL WISE" / 'VALUE WISE" / "ROW WISE"

```python
display(df2.loc['India'])

display(df.loc[df['Country'].str.contains('United')])
df.loc[df['Rank'] >= 100]

df[df['Rank'] >= 100]


df.iloc[3]
```
